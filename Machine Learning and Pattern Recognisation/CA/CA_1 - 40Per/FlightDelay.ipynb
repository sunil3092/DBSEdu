```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("https://drive.google.com/file/d/14N2JcZXNjQVNLccJDk6NZJ-aEZ09RpoB/view?usp=sharing")
df.head()
```

```python
df.isna().sum()
```

```python
df_clear = df.dropna(how='any',axis=0)
df_clear.isna().sum()
```

```python
len(df_clear.index), len(df.index) 
```

```python
columns_to_drop = ["Unnamed: 0","Cancelled","CancellationCode","Diverted","UniqueCarrier","TailNum"]
for value in columns_to_drop:
    if value in  df_clear.columns:
        df_clear.drop(columns_to_drop, axis=1, inplace= True);
df_clear.head()
```

```python
df_clear[["Origin", "Dest"]]
ordinal_enc = OrdinalEncoder()
df_clear[["Origin", "Dest"]] = ordinal_enc.fit_transform(df_clear[["Origin", "Dest"]])
df_clear.head()
```

```python
X = df_clear.drop("ArrDelay", axis=1)
y = df_clear["ArrDelay"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
models = {}
models["DTR"] = DecisionTreeRegressor()
```

```python
model_acc = {}
for key in models:
        for i in range(1):
                model_r = models[key]
                dt = model_r.fit(X_train,y_train) # fit the model
                # Y_pred = dt.predict(X_test) # predict
                a = model_r.score(X_test,y_test) # compute accuracy
        model_acc[key] = a
print(model_acc) 
```
